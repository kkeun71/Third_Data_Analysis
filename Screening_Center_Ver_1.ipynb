### 평일 운영시간 수정

```python
df.loc[df["평일 운영시간"].str.contains('9:00 ~ 17:30'), "평일 운영시간"] = "09:00 ~ 17:30"
df.loc[df["평일 운영시간"].str.contains('9:00 ~ 16:00'), "평일 운영시간"] = "09:00 ~ 16:00"
df.loc[df["평일 운영시간"].str.contains('9:0 ~ 18:0|09:0 ~ 18:00|9:00 ~ 18:00'), "평일 운영시간"] = "09:00 ~ 18:00"
df.loc[df["평일 운영시간"].str.contains('9:00 ~ 17:00'), "평일 운영시간"] = "09:00 ~ 17:00"
df.loc[df["평일 운영시간"].str.contains('9:00 ~ 20:00'), "평일 운영시간"] = "09:00 ~ 20:00"
df.loc[df["평일 운영시간"].str.contains('9:00 ~ 21:00'), "평일 운영시간"] = "09:00 ~ 21:00"
df.loc[df["평일 운영시간"].str.contains('8:30 ~ 17:00'), "평일 운영시간"] = "08:30 ~ 17:00"
df.loc[df["평일 운영시간"].str.contains('8:00 ~ 17:00'), "평일 운영시간"] = "08:00 ~ 17:00"
df.loc[df["평일 운영시간"].str.contains('00:00 ~ 24:00|0:0 ~ 24:00|00:00 ~ 23:59|0:0 ~ 24:0'), "평일 운영시간"] = "24시간"
df.loc[df["평일 운영시간"].str.contains('-'), "평일 운영시간"] = "09:00 ~ 18:00"
```

### 일요일 및 공휴일 운영시간 수정

```python
for i in df.index:
    print(df.loc[i, "일요일/공휴일\n운영시간"])
```

```python
df.loc[df["일요일/공휴일\n운영시간"].str.contains('06:00 ~ 06:00|00:00 ~ 00:00|00:00 ~ 24:00|0:0 ~ 24:0|24:00 ~ 24:00'), "일요일/공휴일\n운영시간"] = "24시간"
df.loc[df["일요일/공휴일\n운영시간"].str.contains('08:0 ~ 23:0'), "일요일/공휴일\n운영시간"] = "08:30 ~ 23:00"
df.loc[df["일요일/공휴일\n운영시간"].str.contains('9:0 ~ 14:0'), "일요일/공휴일\n운영시간"] = "09:00 ~ 14:00"
df.loc[df["일요일/공휴일\n운영시간"].str.contains('-'), "일요일/공휴일\n운영시간"] = "미운영"
df.loc[df["일요일/공휴일\n운영시간"].str.contains('09:0 ~ 13:00'), "일요일/공휴일\n운영시간"] = "09:00 ~ 13:00"
df.loc[df["일요일/공휴일\n운영시간"].str.contains('9:00 ~ 18:00|9:0 ~ 18:0'), "일요일/공휴일\n운영시간"] = "09:00 ~ 18:00"
```
